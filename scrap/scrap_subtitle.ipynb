```python
import requests as re
from bs4 import BeautifulSoup
```

```python
def getSub(mov_title):
    res = re.get(f'http://localhost:3000/api/eng/search?q={"+".join(mov_title.lower().split())}&type=movies')
    movie_res = re.get(res.json()[0]['url'])
    soup = BeautifulSoup(movie_res.text)
    prefix = 'https://www.opensubtitles.org'
    postfix = soup.find(class_='bnone')['href']
    dl_page = re.get(f'{prefix}/en/subtitleserve/sub/{postfix.split("/")[-2]}')
    with open(f'{mov_title}.zip', 'wb') as f:
        f.write(dl_page.content)
```
